```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
wodsDF = pd.read_csv('../data/processed/wodsEngineered.csv')
wodsDF
```

```python
feat_names = list(wodsDF.columns)
feat_names.remove('Name')
feat_names.remove('href')
feat_names.remove('Views')
feat_names.remove('Likes')
feat_names.remove('LikesPerView')
```

```python
y = wodsDF['Likes']
X = wodsDF[feat_names]
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state=42)
```

```python
data_path = '../data/interim'
Xtrain.to_csv(data_path+'wods_Xtrain',index=False)
ytrain.to_csv(data_path+'wods_ytrain',index=False)
Xtest.to_csv(data_path+'wods_Xtest',index=False)
ytest.to_csv(data_path+'wods_ytest',index=False)
```
