```python
bearComplex = wodsDF[np.logical_or(wodsDF['back squat'] == 11111.0, wodsDF['back squat']==44444.0)]
```

```python
numToBear = {}
for name in bearComplex['Name']:
    numToBear[name] = bearComplex[bearComplex['Name']==name]['back squat']/11111.0
```

```python
for loc in range(len(wodsDF)):
    tempName = wodsDF['Name'].iloc[loc]
    if tempName in numToBear.keys():
        wodsDF['back squat'].iloc[loc] = float(numToBear[tempName])
        wodsDF['clean'].iloc[loc] += float(numToBear[tempName])
        wodsDF['push press'].iloc[loc] += float(numToBear[tempName]*2)
        wodsDF['front squat'].iloc[loc] += float(numToBear[tempName])
```
